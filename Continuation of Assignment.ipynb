### Part A

```python
import numpy as np
from scipy.optimize import minimize

def objective(v):
    x,y=v
    return 2*(x-y-3)**2+4*(x+2*y+1)**4
def constraint1(v):
    x,y=v
    return x-y+3
def constraint2(v):
    x,y=v
    return 5-((x+2)**2+(y+1)**2)
constraints=[{"type":"ineq","fun":constraint1},{"type":"ineq","fun":constraint2},]
initial_guess=[0,0]

result=minimize(objective,initial_guess,constraints=constraints)
print("Optimal solution (x,y):",result.x)
print("Minimum value of this function as per x and y values is:",result.fun)
```

### Part B

```python
import numpy as np
from scipy.integrate import quad
def integrand(t):
    x=np.sqrt(3)*np.cos(t)  #root(3)cost
    y=np.sqrt(3)*np.sin(t)  #root(3)sint
    f_xy=x**2+y**4          #f(x,y)
    ds=np.sqrt(3)           #ds/dt is constant
    return f_xy*ds
t_min,t_max=0,2*np.pi

result=quad(integrand,t_min,t_max)
print("Line integral :",result[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
def load_data(df):
    return np.loadtxt(df,delimiter=',')
def initialise_centers(data,K,init_centers=None):
    if init_centers is None:
        return data[np.random.choice(data.shape[0],K,replace=False)]
    return np.array(init_centers)
def initialise_labels(data):
    return np.ones(data.shape[0],dtype=int)
def calculate_distances(data, centers):
    distances=np.linalg.norm(data[:,np.newaxis]-centers,axis=2)
    return distances
def update_labels(distances):
    return np.argmin(distances,axis=1)
def update_centers(data,labels,K):
    centers=np.zeros((K,data.shape[1]))
    for i in range(K):
        cluster_points=data[labels==i]
        if len(cluster_points)>0:
            centers[i]=np.mean(cluster_points,axis=0)
    return centers
def check_termination(labels1,labels2):
    return np.array_equal(labels1,labels2)
def kmeans(data_path:str,K:int,init_centers):
    data=load_data(data_path)    
    centers=initialise_centers(data,K,init_centers)
    labels=initialise_labels(data)
    start_time=time.time()
    while True:
        distances=calculate_distances(data,centers)
        labels_new=update_labels(distances)
        centers=update_centers(data,labels_new,K)
        if check_termination(labels,labels_new):
            break
        else:
            labels=labels_new
    end_time=time.time()
    return centers,labels,end_time-start_time
def visualise(data_path,labels,centers):
    data=load_data(data_path)
    plt.scatter(data[:,0],data[:,1],c=labels,s=50,cmap='viridis')
    plt.scatter(centers[:,0],centers[:,1],c='black',s=200,alpha=0.5)
    plt.title('K-means clustering')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.savefig('kmeans.png')
    return plt
data_path=r"C:\Users\Mahima\Downloads\spice_locations.txt"
K = 3  # Number of clusters
init_centers = None  # Use random initialization of centers

# Run K-means
centers, labels, time_taken = kmeans(data_path, K, init_centers)

# Visualize the results
visualise(data_path, labels, centers)

# To show the plot if needed:
# plt.show()
```
